```python
# By Zain Nasrullah
import pandas as pd
from pandas import DataFrame
from pylab import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Go to https://archive.ics.uci.edu/ml/datasets/Wine+Quality and read about the
# dataset

# This will set data_url to the URL of the dataset so you can download it
data_url = ("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv")
```

```python
# Download the data
red_wine_data = pd.read_csv(data_url, header=0, sep=";")
red_wine_data.head(n=20)
```

```python
# Question 1: Use the DataFrame .describe() method to show summary statistics for
# the red wine dataset

red_wine_data.describe()
```

```python
# Question 2: Normalize the data so all attributes (columns) are on the same scale
# One way to do this is to scale each column so the mean becomes zero and
# the data is in units of standard deviation. (See
# https://en.wikipedia.org/wiki/Normalization_(statistics))
# Alternatively you can scale the data so all columns are on a scale of 0 to 1.
# You can do either for this exercise.

# Used feature scaling below as per the wikipedia article to bring all data in the range [0,1]
normalized_red_wine_data = (red_wine_data - red_wine_data.min()) / (red_wine_data.max() - red_wine_data.min())
normalized_red_wine_data.head(n=20)
```

```python
# Question 3: Test the data using .isnull() to see if there are any missing data values.
# How many missing values are there?

# No missing values in the data below
red_wine_data.isnull().sum()
```

```python
# This will show a boxplot of the data.  We will explain boxplots a little
# later in the course.
boxplot(normalized_red_wine_data.values)
show()
```

```python
# This is a better boxplot
matplotlib.rcParams['figure.figsize'] = (22.0, 8.0)
normalized_red_wine_data.boxplot(return_type='axes')
```

```python
# This will show a scatter plot of all possible pairs of attributes
pd.scatter_matrix(red_wine_data, alpha=0.2, figsize=(20,20), diagonal='kde')
show()

# Question 4: Pick out two examples of attributes that appear to be related to each other.

# A few examples of potentially related attributes:
print('''
1) Fixed Acidity & Cirtric Acid as per the diagram's below appear to have a proportional relationship 
2) Fixed Acidity and pH also appear to have an inversely proportional relationship
3) Fixed Acidity & Density also appear to have a proportional relationship
''')
```

```python
density2 = normalized_red_wine_data['density'].tolist()
acidity2 = normalized_red_wine_data['fixed acidity'].tolist()
pH2 = normalized_red_wine_data['pH'].tolist()

plt.plot(acidity2,pH2,'bo',label='pH')
plt.plot(acidity2, density2, 'ro', label = 'Density')
plt.legend(loc = "upper left")

plt.xlabel("Fixed Acidity")
plt.show()
```
